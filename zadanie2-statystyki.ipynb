```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats
from scipy.stats import norm

```

```python
r = np.random.normal(loc=3, scale=1, size=100)
```

```python
def gauss(x, sigma=1, mu=0):
    return stats.norm.pdf(x, mu, sigma)
x = np.linspace(-6, 7, 100)
plt.figure(figsize=(10, 5))
plt.hist(r, density=True, bins='auto', histtype='stepfilled', alpha=0.2)
plt.plot(x, gauss(x, sigma=1, mu=3), label='odchylenie standardowe = 1, średnia = 3')
plt.legend(loc='upper left')
plt.show()
```

Średnia

```python
mean_r = round(np.mean(r),2)
mean_r
```

```python
median_r = round(np.median(r),2)
median_r
```

```python
mode_r = stats.mode(r)
mode_r = round(mode_r[0],2)
mode_r
```

<b>Kwartyle</b>

```python
q0 = np.quantile(r, 0.0)
q1 = np.round(np.quantile(r, 0.25))
q3 = np.round(np.quantile(r, 0.75))
q4 = np.round(np.quantile(r, 1.0))
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median_r}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')
```

<b>Zakres</b>

```python
range_r = max(r) - min(r)
round(range_r, 2)
```

<b>Rozstęp międzykwartylowy - IQR</b>

```python
iqr = q3 - q1
iqr
```

<b>Wykres pudełkowy</b>

```python
plt.boxplot(r)
plt.show()
```

<b>Wariancja</b>

```python
variance_r = round(np.var(r, ddof=1),2)
variance_r
```

<b>Odchylenie standardowe</b>

```python
print(variance_r**(1/2))
```

```python
plt.figure(figsize=(10, 5))
plt.hist(r, density=True, bins='auto', histtype='stepfilled', alpha=0.2)
plt.plot(x, gauss(x, sigma=1, mu=3), label='odchylenie standardowe = 1, średnia = 3')
plt.axvline(x = mean_r, color='yellow', lw=6, label = 'Średnia')
plt.axvline(x = median_r, color='green', lw=4, label = 'Mediana')
plt.axvline(x = mode_r, color='red', lw=4, label = 'Moda')
plt.axvline(x = q0, color='black', lw=2, label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')
plt.legend(loc='upper left')
plt.show()
```
